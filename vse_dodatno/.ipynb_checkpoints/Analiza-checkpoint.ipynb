```python
import pandas as pd

import matplotlib
%matplotlib inline
import numpy as np

from matplotlib import pyplot as plot
```

```python
zapis = pd.read_csv('podatki_za_obdelavo.csv', encoding = "utf-8", sep=';')
zapis['Birthday'] = pd.to_datetime(zapis['Birthday'])
pd.parse_dates=['Birthday'] #kaj sploh delajo te funkcije?
zapis.index = zapis['Birthday']
```

```python
zapis.sort_values(by=['Birthday'], ascending=True)
```

```python
zapis.iloc[1]
```

```python
#zapis(parse_dates=['Birthday'])
zapis[:1000].plot(x='Birthday', y='Rank', kind='scatter', title='Razmerje')
```

```python
drzava = zapis.groupby('Birthday')
drzava = drzava.count()
```
